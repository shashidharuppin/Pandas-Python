# Reading Excel file

```python
df = pd.read_excel("C:\\Users\\UPPI\\Documents\\Netzwerk\\ML Assign#1\\stock_data.xlsx","stock_data")
df # reading the excel file with the file name and next sheet name
```

```python
#Cleaning up the data using the functions for each cell of the df

def convert_people_cell(cell):
    if cell=="n.a.":
        return 'Sam Walton'
    return cell

def convert_price_cell(cell):
    if cell=="n.a.":
        return 50
    return cell
    
def convert_revenue_cell(cell):
    if cell < 0:
        return 0
    return cell

def convert_eps_cell(cell):
    if cell=="not available":
        return None
    return cell

df = pd.read_excel("C:\\Users\\UPPI\\Documents\\Netzwerk\\ML Assign#1\\stock_data.xlsx","stock_data", converters= {
        'people': convert_people_cell,
        'price': convert_price_cell,
        'revenue':convert_revenue_cell,
        'eps': convert_eps_cell
    })
df
```

# Writing to excel file

```python
df.to_excel("C:\\Users\\UPPI\\Documents\\Netzwerk\\ML Assign#1\\new.xlsx", sheet_name="stocks", index=False)
```
