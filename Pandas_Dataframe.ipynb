```python
name = ['Shakti', 'Hemanka', 'Rajeshwari', 'Akshaya', 'Fenil']
domain = ['NLP', 'Reinforcement Learning', 'Computer Vision', 'Analytics', 'Automation']
experience = [2, 3, 4, 5, 6]
dct = {'Name': name, 'Domain': domain, 'Experience': experience} #Creating a dictionary 
```

```python
weather_data = [
    ('1/1/2017',32,6,'Rain'),
    ('1/2/2017',35,7,'Sunny'),
    ('1/3/2017',28,2,'Snow')
]
df3 = pd.DataFrame(data=weather_data, columns=['day','temperature','windspeed','event']) #Creating data frame from list of tuples
df3
```

```python
#Creating data frame from list of dictionaries

weather_data = [
    {'day': '1/1/2017', 'temperature': 32, 'windspeed': 6, 'event': 'Rain'},
    {'day': '1/2/2017', 'temperature': 35, 'windspeed': 7, 'event': 'Sunny'},
    {'day': '1/3/2017', 'temperature': 28, 'windspeed': 2, 'event': 'Snow'},
    
]
df4 = pd.DataFrame(data=weather_data, columns=['day','temperature','windspeed','event'])
df4
```
