# Pivot in Pandas

```python
import pandas as pd
import numpy as np
df = pd.read_csv("C:\\Users\\UPPI\\Documents\\Netzwerk\\ML Assign#1\\weather.csv")
df
```

```python
df.pivot(index='date',columns = 'city')
```

```python
df.pivot(index='date',columns = 'city',values = 'temperature')
```

```python
df.pivot(index='humidity',columns='city')
```

# Pivot Table in Pandas

```python
df = pd.read_csv("C:\\Users\\UPPI\\Documents\\Netzwerk\\ML Assign#1\\weather2.csv")
df
```

```python
df.pivot_table(index='date',columns='city')
```

```python
df.pivot_table(index='date',columns='city',aggfunc='sum')
```
